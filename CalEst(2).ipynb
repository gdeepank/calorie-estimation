```python
# list of label names along with 'BG' (background) class

class_names=['BG','egg', 'mooncake', 'doughnut', 'bun', 'apple', 'pear', 'mango', 
             'litchi', 'qiwi', 'grape', 'plum', 'banana', 'fired_dough_twist', 
             'lemon', 'peach', 'tomato', 'sachima', 'bread', 'orange', 'coin']
```

```python
# importing few helpful libraries

from google.colab import files
import pixellib
from pixellib.instance import custom_segmentation
import matplotlib.pyplot as plt
from matplotlib import image as im
```

```python
# uploading model weights for inference

files.upload()
```

```python
# loading model weights for inference

segment_image = custom_segmentation()
segment_image.inferConfig(num_classes=20, class_names=class_names)
segment_image.load_model("/content/mask_rcnn_model.003-0.312321.h5")
```

```python
# performing inference on a sample test set image

segment_image.segmentImage("/content/ECUSTFD-resized-/JPEGImages/lemon001T(13).JPG", show_bboxes=True, output_image_name="output1.png")
```

```python
# visualizing predictions on sample test set image 

img = im.imread('output1.png')
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
# extracting segmented objects - 1

segment_image.segmentImage("/content/ECUSTFD-resized-/JPEGImages/orange011S(6).JPG", show_bboxes=True, output_image_name="output2.png",
extract_segmented_objects=True, save_extracted_objects=True)
```

```python
# visualizing segmented objects - 1

img = im.imread('output2.png')
seg1 = im.imread('segmented_object_1.png')
seg2 = im.imread('segmented_object_2.png')
plt.axis('off')
plt.imshow(img)
plt.show()
plt.axis('off')
plt.imshow(seg1)
plt.show()
plt.axis('off')
plt.imshow(seg2)
plt.show()
```

```python
# extracting segmented objects - 2

segment_image.segmentImage("/content/ECUSTFD-resized-/JPEGImages/sachima002T(10).JPG", show_bboxes=True, output_image_name="output3.png",
extract_segmented_objects=True, save_extracted_objects=True)
```

```python
# visualizing segmented objects - 2

img = im.imread('output3.png')
seg1 = im.imread('segmented_object_1.png')
seg2 = im.imread('segmented_object_2.png')
plt.axis('off')
plt.imshow(img)
plt.show()
plt.axis('off')
plt.imshow(seg1)
plt.show()
plt.axis('off')
plt.imshow(seg2)
plt.show()
```
